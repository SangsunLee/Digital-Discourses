## Word frequencies

```python
# Just for analyzing the general corpus
word_freq = nltk.FreqDist(words)
print(word_freq.most_common(50))
```

```python
for word, freq in word_freq.most_common(50):
    percentage = word_freq.freq(word) * 100 # we would get the same result with this as well: freq / len(words) * 100
    # \t is tabulation character, str is function for converting numbers to strings
    print(word + '\t' + str(freq) + '\t' + str(round(percentage, 2)) + '%')
```

## Part of Speech tagging

The codes are listed at the [Penn Treebank Project](http://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html). Below is the full table:

<div style="float: center; padding: 10px 0 10px 0;"><img style="float: center;" src="https://thottingal.in/wp-content/uploads/2019/09/PENN-treebank-tagset.png" width="80%" height="80%" align=center /></div>

```python
# Doing POS tagging
tagged = nltk.pos_tag(words)
print(tagged)
```

```python
# Examining frequencies and distribution for a specific part-of-speech
tags_normalized = []
for token, tag in tagged:
    if tag[0].isalpha():
        tags_normalized.append(tag[0:2])
    else:
        tags_normalized.append('.')

tags_normalized_freq = nltk.FreqDist(tags_normalized)
tags_normalized_freq.tabulate()
tags_normalized_freq.plot()
```

```python
# In this project, listing the most frequently used adjectives are very important
# Because it represent parrents' emotions and evaluations about Peppa Pig
# Figuring what 100 adjectives appear most
adjectives = [token for token, tag in tagged if tag[:2] == 'JJ']
adjectives_freqs = nltk.FreqDist(adjectives)
plt.figure(figsize=(16, 4))
adjectives_freqs.plot(100)
```

## Stemming

```python
# Stemming can reduce the vocabulary and focus on the sense or the corpus
from nltk.stem.snowball import SnowballStemmer
snowball = SnowballStemmer('english')
```

```python
# Counting the number of unique stems
stems = [snowball.stem(word) for word in words]
unique_stems = len(set(stems))

total_words = len(words)
unique_words = len(set(words))

print('Number of unique stems in the text:', unique_stems, '(' + str(round(unique_stems / total_words * 100, 1))+ '%)')
```

```python
# Printing the whole stems
print(stems)
```

```python
# Printing out result of stems. However, the result didn't fit this project than 100 adjectives did above.
word_freq = nltk.FreqDist(stems)
print(word_freq.most_common(100))
```

## Lemmatization

```python
# Lemmatization uses a vocabulary and morphological analysis of words and is normally aiming to return the base or dictionary form of a word, which is known as the lemma.
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python
# Counting the number of unique stems
lemmas = [wnl.lemmatize(word) for word in words]
unique_lemmas = len(set(lemmas))

print('Number of unique lemmas in the text:', unique_lemmas, '(' + str(round(unique_lemmas / total_words * 100, 1))+ '%)')
```

```python
# Printing the whole stems
print(lemmas)
```

```python
# Printing out result of stems. However, the result didn't fit this project than 100 adjectives did above.
word_freq = nltk.FreqDist(lemmas)
print(word_freq.most_common(100))
```
